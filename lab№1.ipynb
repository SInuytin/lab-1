##### <u>Часть 2. Численный метод</u>
    Дана последовательность:
    
$$x_n = \frac{5n - 7}{2n + 5} \cdot \cos{\frac{2 + (-1)^n}{6}\pi}$$

1. Постороим график первых 100 точек последовательности $x_n$
2. Отметим на графике найденные аналитически $\sup x_n, \inf x_n, \limsup x_n \liminf x_n$

```python
%matplotlib inline 
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def x(n): #функция вычисления n члена последовательности
    return (5*n - 7)/(2*n + 5) * math.cos((2+pow(-1,n))/6*math.pi)

x = np.vectorize(x)
n = np.arange(0,100)
plt.figure().set_figwidth(18)
plt.plot(n, x(n), 'or--', linewidth=0.9)
inf = 0
sup = 5*math.sqrt(3)/4
a = [0,100]
b = [inf,inf]
c = [sup,sup]
plt.plot(a,b,'b', linewidth=1)
plt.plot(a,c,'b', linewidth=1)
```

3. Из графика наглядно видно две подпоследовательности $2k$ и $2k+1$, где $k \in N$
4. Выделим сходящиеся последовательности и отметим ее точки на графике другим цветом.
5. Последовательность $x_{2n}$ сходится к $\frac{5\sqrt{3}}{4}$, а $x_{2n+1}$ сходится к $0$.

```python
plt.figure().set_figwidth(18)
a = np.arange(0,100)
plt.plot(a,x(a), 'r--', linewidth=0.9, label = "Изнач. послед.")
#красный цветом обозначена оригинальная последовательность x_n

n = np.arange(0,100,2)
plt.plot(n, x(n), 'og', linewidth=0.9, label = "2k. подпослед.")
#зеленым цветом обозначена подпоследовательность четных элементов

m = np.arange(1,100,2)
plt.plot(m,x(m), 'oy', linewidth=0.9, label = "2k+1. подпослед.")
#желтым цветом обозначена подпоследовательность нечетных элементов
a = [0,100]
b = [0,0]
c = [sup,sup]
plt.plot(a,b,'b', linewidth=1,label = "Нижняя граница")
plt.plot(a,c,'c', linewidth=1,label = "Верхняя граница")
plt.legend(loc="upper left")
plt.show
```

5. Выберем подпоследовательность $x_{2k+1}$, которая сходится.
6. По данному $\epsilon > 0$ найдем $n_0,$ начиная с которого все члены подпоследовательности $x_{2n+1}$  попадают в $\epsilon$-окресность предела.
7. Построем график $x_{2n+1}$ начиная с найденного $n_0$ (1000 точек), отметим горизантальной линией значение предела

```python
def x(n): #функция вычисления n члена последовательности
    return (5*n - 7)/(2*n + 5) * math.cos((2+pow(-1,n))/6*math.pi)
n0 = -1
eps = 0.1#эпсилон возьмем небольшое для большей наглядности
for i in range(1,10000,2):
    if sup - x(i) < eps:
        n0 = i
        break
print(n0,sup,x(n0))#найденные значение n0, супремума и x(n_0)
```

```python
x = np.vectorize(x)
n = np.arange(n0,n0+1000,2)
plt.figure().set_figwidth(15)
plt.grid()
plt.plot(n, x(n), 'r--', linewidth=0.9)
a = [n0,n0+1000]
b = [sup,sup]
plt.plot(a,b,'b', linewidth=1)
```

8. Для исходной последовательности $x_n,$ найдем $n_0$ для нижнего и верхнего предела.
9. Проверим выполнения критерия точной грани.

```python
def x(n): #функция вычисления n члена последовательности
    return (5*n - 7)/(2*n + 5) * math.cos((2+pow(-1,n))/6*math.pi)
n0 = -1
uo = -1
eps = 0.1 #эпсилон возьмем небольшое для большей наглядности
for i in range(0,1000):# Находим N_0 для sup(x_n)
    y = x(i)
    if sup - y < eps:
        u0 = i
        break
    if inf - y < eps:
        n0 = i
for i in range(2,1000):# Находим Находим N_0 для inf(x_n)
    if inf - y < eps:
        n0 = i
        break
x = np.vectorize(x)
n = np.arange(0,100)
plt.figure().set_figwidth(15)
plt.grid()
plt.plot(u0, x(u0), 'oy', linewidth=0.9)
plt.plot(n0, x(n0), 'om', linewidth=0.9)
plt.plot(n, x(n), 'r--', linewidth=0.9)
a = [0,100]
b = [sup,sup]
c = [inf,inf]
plt.plot(a,b,'b', linewidth=1)
plt.plot(a,c,'c', linewidth=1)
print("N_0 для sup(x_n):",u0)
print("N_0 для inf(x_n):",n0)
```
